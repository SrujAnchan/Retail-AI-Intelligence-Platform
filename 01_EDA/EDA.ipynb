````python
# =============================================================================
# RETAIL DATASET — COMPLETE DEEP ANALYSIS & BI ARCHITECTURE PLANNING
# Senior Data Engineer + Analytics Engineer + BI Architect Assessment
# Run in Google Colab after uploading all 6 CSV files
# =============================================================================

# ─── INSTALL / IMPORTS ───────────────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
import os
import re
from IPython.display import display, Markdown

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 50)
pd.set_option("display.max_colwidth", 80)
pd.set_option("display.float_format", lambda x: f"{x:,.4f}")

# Consistent colour palette
PALETTE = {
    "primary":   "#2563EB",
    "secondary": "#7C3AED",
    "accent":    "#F59E0B",
    "danger":    "#EF4444",
    "success":   "#10B981",
    "neutral":   "#6B7280",
    "bg":        "#F8FAFC",
}
sns.set_theme(style="whitegrid", palette="muted", font_scale=1.05)

def hdr(text, level=2):
    display(Markdown("#" * level + " " + text))

def info(text):
    display(Markdown(f"> ℹ️  {text}"))

def warn(text):
    display(Markdown(f"> ⚠️  **WARNING:** {text}"))

def ok(text):
    display(Markdown(f"> ✅  {text}"))

# =============================================================================
# STEP 1 — LOAD ALL DATASETS
# =============================================================================
hdr("STEP 1 — LOAD ALL DATASETS", 1)
info("Loading all 6 CSV files with automatic encoding & delimiter detection.")

FILE_PATHS = {
    "customers":     "Customers.csv",
    "data_dict":     "Data_Dictionary.csv",
    "exchange_rates":"Exchange_Rates.csv",
    "products":      "Products.csv",
    "sales":         "Sales.csv",
    "stores":        "Stores.csv",
}

def smart_load(path):
    """Try UTF-8, then latin-1; detect separator automatically."""
    for enc in ["utf-8", "latin-1", "cp1252"]:
        try:
            df = pd.read_csv(path, encoding=enc, sep=None, engine="python",
                             on_bad_lines="warn")
            return df, enc
        except Exception:
            continue
    raise ValueError(f"Cannot read {path}")

FRAMES = {}
LOAD_META = []
for name, path in FILE_PATHS.items():
    if not os.path.exists(path):
        warn(f"`{path}` not found — skipping.")
        continue
    df, enc = smart_load(path)
    FRAMES[name] = df
    LOAD_META.append({
        "Table":         name,
        "File":          path,
        "Rows":          len(df),
        "Columns":       len(df.columns),
        "Encoding":      enc,
        "Memory (KB)":   round(df.memory_usage(deep=True).sum() / 1024, 1),
    })

load_summary = pd.DataFrame(LOAD_META)
hdr("Dataset Overview Table")
display(load_summary.style.set_caption("All Loaded Tables")
        .bar(subset=["Rows"], color=PALETTE["primary"])
        .bar(subset=["Memory (KB)"], color=PALETTE["accent"]))

# =============================================================================
# STEP 2 — COLUMN-LEVEL ANALYSIS (every table)
# =============================================================================
hdr("STEP 2 — COLUMN-LEVEL ANALYSIS", 1)
info("For each table: dtypes, nulls, uniques, sample values, string lengths, suspicious values.")

def col_profile(df, table_name):
    hdr(f"Table: `{table_name.upper()}`", 3)
    rows = []
    for col in df.columns:
        s = df[col]
        null_cnt  = s.isna().sum()
        null_pct  = round(null_cnt / len(s) * 100, 2)
        unique    = s.nunique()
        dup_cnt   = len(s) - unique - null_cnt

        # min / max for numeric/date
        try:
            mn = s.min(); mx = s.max()
        except Exception:
            mn = mx = "—"

        # string lengths
        if s.dtype == object:
            lens = s.dropna().astype(str).str.len()
            str_info = f"len {int(lens.min())}–{int(lens.max())} (mean {lens.mean():.1f})"
        else:
            str_info = "—"

        # sample values (top 3 non-null)
        sample = s.dropna().unique()[:3].tolist()

        rows.append({
            "Column":       col,
            "Dtype":        str(s.dtype),
            "Nulls":        null_cnt,
            "Null%":        null_pct,
            "Uniques":      unique,
            "Dups (est)":   max(dup_cnt, 0),
            "Min":          mn,
            "Max":          mx,
            "String Info":  str_info,
            "Sample Values":str(sample),
        })

    profile_df = pd.DataFrame(rows)
    display(profile_df.style
            .highlight_between(subset=["Null%"], left=20, right=100,
                               color="#FEE2E2")
            .set_caption(f"Column Profile — {table_name}"))

    # suspicious checks
    issues = []
    for col in df.columns:
        s = df[col]
        if s.dtype == object:
            if s.str.contains(r"^\s+|\s+$", na=False).any():
                issues.append(f"**{col}**: trailing/leading spaces detected")
            if s.str.lower().nunique() < s.nunique():
                issues.append(f"**{col}**: inconsistent casing")
            empty_str = (s == "").sum()
            if empty_str:
                issues.append(f"**{col}**: {empty_str} empty strings")
        if pd.api.types.is_numeric_dtype(s):
            neg = (s < 0).sum()
            if neg:
                issues.append(f"**{col}**: {neg} negative values")

    if issues:
        warn(f"Suspicious findings in `{table_name}`:\n" + "\n".join(f"- {i}" for i in issues))
    else:
        ok(f"No obvious suspicious values in `{table_name}`.")

for name, df in FRAMES.items():
    col_profile(df, name)

# =============================================================================
# STEP 3 — TABLE UNDERSTANDING (business interpretation)
# =============================================================================
hdr("STEP 3 — TABLE BUSINESS INTERPRETATION", 1)

TABLE_BUSINESS = {
    "customers": {
        "purpose":  "Master dimension of all buyers. Used to segment by demographics, geography, loyalty.",
        "role":     "DIM_CUSTOMER in star schema. Central to RFM, CLV, churn analyses.",
        "key_cols": "CustomerKey / CustomerID, City, State/Country, Gender, Birth Date",
    },
    "data_dict": {
        "purpose":  "Metadata catalog describing all columns across tables.",
        "role":     "Documentation layer. Should drive column naming standards in dbt staging models.",
        "key_cols": "Table, Field, Description, Data Type",
    },
    "exchange_rates": {
        "purpose":  "Currency conversion reference for multi-currency sales normalization.",
        "role":     "Helper dimension / snapshot table. Applied in dbt intermediate layer to convert to base currency.",
        "key_cols": "Date, Currency, Exchange Rate",
    },
    "products": {
        "purpose":  "Product master with pricing, category, and cost information.",
        "role":     "DIM_PRODUCT. Drives product performance, margin, and category analyses.",
        "key_cols": "ProductKey, Category, Subcategory, Unit Price, Unit Cost",
    },
    "sales": {
        "purpose":  "Transactional fact table — the core of all revenue reporting.",
        "role":     "FACT_SALES. Joins to all dimensions. Primary table for KPIs.",
        "key_cols": "Order Number, Line Item, Order Date, CustomerKey, StoreKey, ProductKey, Quantity, Currency, Unit Price",
    },
    "stores": {
        "purpose":  "Store/outlet master — geographic and operational metadata.",
        "role":     "DIM_STORE. Enables regional, city, country-level rollups.",
        "key_cols": "StoreKey, Country, State, Area (sqm), Open Date",
    },
}

for tbl, meta in TABLE_BUSINESS.items():
    if tbl not in FRAMES:
        continue
    hdr(f"`{tbl.upper()}`", 3)
    display(Markdown(f"""
| Attribute | Detail |
|-----------|--------|
| **Business Purpose** | {meta['purpose']} |
| **BI Role** | {meta['role']} |
| **Key Columns** | {meta['key_cols']} |
| **Row Count** | {len(FRAMES[tbl]):,} |
"""))

# =============================================================================
# STEP 4 — PRIMARY KEY DETECTION
# =============================================================================
hdr("STEP 4 — PRIMARY KEY DETECTION", 1)
info("Testing every column (and composite pairs) for uniqueness to find PK candidates.")

def test_pk(df, col_combo):
    subset = df[list(col_combo)].dropna()
    total  = len(subset)
    uniq   = subset.drop_duplicates().shape[0]
    pct    = round(uniq / total * 100, 2) if total else 0
    return pct, total - uniq

PK_RESULTS = {}
for name, df in FRAMES.items():
    hdr(f"Table: `{name.upper()}`", 3)
    rows = []
    for col in df.columns:
        pct, dups = test_pk(df, [col])
        rows.append({"Column(s)": col, "Unique%": pct, "Duplicates": dups,
                     "Verdict": "✅ Strong PK" if pct == 100 else ("⚠️ Weak" if pct > 90 else "❌ Not PK")})
    pk_df = pd.DataFrame(rows).sort_values("Unique%", ascending=False)
    display(pk_df.style.set_caption(f"PK Analysis — {name}"))
    best = pk_df[pk_df["Unique%"] == 100]
    if not best.empty:
        ok(f"Likely PK for `{name}`: `{best.iloc[0]['Column(s)']}`")
        PK_RESULTS[name] = best.iloc[0]["Column(s)"]
    else:
        warn(f"No single-column PK found for `{name}`. Consider composite key.")

# =============================================================================
# STEP 5 — FOREIGN KEY & RELATIONSHIP DETECTION
# =============================================================================
hdr("STEP 5 — FOREIGN KEY & RELATIONSHIP DETECTION", 1)
info("Matching key columns across tables to detect FK relationships and orphan records.")

EXPECTED_RELS = [
    ("sales",    "CustomerKey", "customers",    "CustomerKey"),
    ("sales",    "ProductKey",  "products",     "ProductKey"),
    ("sales",    "StoreKey",    "stores",       "StoreKey"),
]

rel_rows = []
for (fact_tbl, fact_col, dim_tbl, dim_col) in EXPECTED_RELS:
    if fact_tbl not in FRAMES or dim_tbl not in FRAMES:
        continue
    fact_df = FRAMES[fact_tbl]
    dim_df  = FRAMES[dim_tbl]

    # fuzzy column match
    fc = next((c for c in fact_df.columns if fact_col.lower() in c.lower()), None)
    dc = next((c for c in dim_df.columns  if dim_col.lower()  in c.lower()), None)
    if fc is None or dc is None:
        warn(f"Column not found: {fact_tbl}.{fact_col} → {dim_tbl}.{dim_col}")
        continue

    fact_keys = set(fact_df[fc].dropna().unique())
    dim_keys  = set(dim_df[dc].dropna().unique())
    orphans   = fact_keys - dim_keys
    match_pct = round((len(fact_keys - orphans) / len(fact_keys)) * 100, 2) if fact_keys else 0

    rel_rows.append({
        "Relationship":   f"{fact_tbl}.{fc} → {dim_tbl}.{dc}",
        "Fact Keys":      len(fact_keys),
        "Dim Keys":       len(dim_keys),
        "Match%":         match_pct,
        "Orphans":        len(orphans),
        "Cardinality":    "Many-to-One",
        "Status":         "✅ OK" if match_pct == 100 else "⚠️ Orphans exist",
    })

rel_df = pd.DataFrame(rel_rows)
display(rel_df.style.set_caption("FK Relationship Audit"))

# =============================================================================
# STEP 6 — ER DIAGRAM (text + Mermaid)
# =============================================================================
hdr("STEP 6 — ER DIAGRAM", 1)

hdr("Text-Based ER Diagram", 3)
ER_TEXT = """
┌─────────────────────┐        ┌─────────────────────────┐
│     CUSTOMERS        │        │         SALES            │
│─────────────────────│        │─────────────────────────│
│ PK: CustomerKey     │◄───────│ FK: CustomerKey          │
│     Gender          │        │ FK: ProductKey           │
│     Name            │        │ FK: StoreKey             │
│     City/State      │        │     OrderNumber (PK)     │
│     Birthday        │        │     LineItem             │
└─────────────────────┘        │     OrderDate            │
                                │     Quantity             │
┌─────────────────────┐        │     CurrencyCode         │
│     PRODUCTS         │        │     UnitPrice            │
│─────────────────────│        └─────────────────────────┘
│ PK: ProductKey      │◄───────────────┘
│     ProductName     │
│     Category        │        ┌─────────────────────────┐
│     Subcategory     │        │     EXCHANGE_RATES       │
│     UnitPrice       │        │─────────────────────────│
│     UnitCost        │        │     Date                 │
└─────────────────────┘        │     Currency             │
                                │     Exchange             │
┌─────────────────────┐        └─────────────────────────┘
│     STORES           │
│─────────────────────│        ┌─────────────────────────┐
│ PK: StoreKey        │◄───────│     DATA_DICTIONARY      │
│     Country/State   │        │─────────────────────────│
│     Area (sqm)      │        │     Table                │
│     OpenDate        │        │     Field                │
└─────────────────────┘        │     Description          │
                                └─────────────────────────┘
"""
print(ER_TEXT)

hdr("Mermaid ER Diagram Code", 3)
MERMAID = """
```mermaid
erDiagram
    CUSTOMERS {
        int CustomerKey PK
        string Gender
        string Name
        string City
        string StateCode
        string Country
        string Continent
        date Birthday
    }
    PRODUCTS {
        int ProductKey PK
        string ProductName
        string Brand
        string Color
        float UnitCostUSD
        float UnitPriceUSD
        string SubcategoryCode
        string Subcategory
        string CategoryCode
        string Category
    }
    STORES {
        int StoreKey PK
        string Country
        string State
        int SquareMeters
        date OpenDate
    }
    EXCHANGE_RATES {
        date Date
        string Currency
        float Exchange
    }
    SALES {
        string OrderNumber PK
        int LineItem PK
        date OrderDate
        date DeliveryDate
        int CustomerKey FK
        int StoreKey FK
        int ProductKey FK
        int Quantity
        string CurrencyCode FK
    }

    CUSTOMERS  ||--o{ SALES : "places"
    PRODUCTS   ||--o{ SALES : "appears in"
    STORES     ||--o{ SALES : "fulfilled by"
    EXCHANGE_RATES }o--|| SALES : "currency ref"
```
"""
display(Markdown(MERMAID))

# =============================================================================
# STEP 7 — DATA QUALITY AUDIT
# =============================================================================
hdr("STEP 7 — DATA QUALITY AUDIT", 1)
info("Full quality sweep: duplicates, nulls, outliers, impossible dates, negative values, broken FKs.")

quality_issues = {"HIGH": [], "MEDIUM": [], "LOW": []}

for name, df in FRAMES.items():
    # duplicates
    dup_rows = df.duplicated().sum()
    if dup_rows > 0:
        quality_issues["HIGH"].append(f"`{name}`: {dup_rows} fully duplicate rows")

    # high null %
    for col in df.columns:
        pct = df[col].isna().mean() * 100
        if pct > 50:
            quality_issues["HIGH"].append(f"`{name}.{col}`: {pct:.1f}% nulls")
        elif pct > 20:
            quality_issues["MEDIUM"].append(f"`{name}.{col}`: {pct:.1f}% nulls")
        elif pct > 5:
            quality_issues["LOW"].append(f"`{name}.{col}`: {pct:.1f}% nulls")

    # numeric negatives
    for col in df.select_dtypes(include=[np.number]).columns:
        neg = (df[col] < 0).sum()
        if neg:
            severity = "HIGH" if "price" in col.lower() or "quantity" in col.lower() else "MEDIUM"
            quality_issues[severity].append(f"`{name}.{col}`: {neg} negative values")

    # date range checks
    for col in df.columns:
        if "date" in col.lower():
            try:
                parsed = pd.to_datetime(df[col], errors="coerce")
                future = (parsed > pd.Timestamp.now()).sum()
                if future:
                    quality_issues["MEDIUM"].append(f"`{name}.{col}`: {future} future dates")
                very_old = (parsed < pd.Timestamp("1900-01-01")).sum()
                if very_old:
                    quality_issues["HIGH"].append(f"`{name}.{col}`: {very_old} dates before 1900")
            except Exception:
                pass

for level, items in quality_issues.items():
    colour = {"HIGH": "🔴", "MEDIUM": "🟡", "LOW": "🟢"}[level]
    hdr(f"{colour} {level} RISK ISSUES ({len(items)} found)", 3)
    if items:
        for item in items:
            display(Markdown(f"- {item}"))
    else:
        ok("No issues at this level.")

# =============================================================================
# STEP 8 — TIME ANALYSIS
# =============================================================================
hdr("STEP 8 — TIME ANALYSIS", 1)

for name, df in FRAMES.items():
    date_cols = [c for c in df.columns if "date" in c.lower()]
    if not date_cols:
        continue
    hdr(f"Table: `{name.upper()}` — Date Columns: {date_cols}", 3)
    for col in date_cols:
        parsed = pd.to_datetime(df[col], errors="coerce")
        valid  = parsed.dropna()
        if valid.empty:
            warn(f"`{col}` has no parseable dates.")
            continue
        info(f"`{col}` | Range: {valid.min().date()} → {valid.max().date()} | "
             f"Nulls: {parsed.isna().sum()} | Coverage: {len(valid):,} rows")

# Sales date distribution chart
if "sales" in FRAMES:
    df_s = FRAMES["sales"].copy()
    date_col = next((c for c in df_s.columns if "order" in c.lower() and "date" in c.lower()), None)
    if date_col:
        df_s["parsed_date"] = pd.to_datetime(df_s[date_col], errors="coerce")
        df_s["YearMonth"]   = df_s["parsed_date"].dt.to_period("M")
        monthly = df_s.groupby("YearMonth").size().reset_index(name="Orders")
        monthly["YearMonth"] = monthly["YearMonth"].astype(str)

        fig, axes = plt.subplots(1, 2, figsize=(16, 5))
        fig.suptitle("Sales — Time Distribution", fontsize=14, fontweight="bold")

        axes[0].bar(monthly["YearMonth"], monthly["Orders"],
                    color=PALETTE["primary"], alpha=0.85, edgecolor="white")
        axes[0].set_title("Monthly Order Volume")
        axes[0].set_xlabel("Month"); axes[0].set_ylabel("Orders")
        axes[0].tick_params(axis="x", rotation=45)

        yearly = df_s.groupby(df_s["parsed_date"].dt.year).size()
        axes[1].bar(yearly.index.astype(str), yearly.values,
                    color=PALETTE["secondary"], alpha=0.85, edgecolor="white")
        axes[1].set_title("Annual Order Volume")
        axes[1].set_xlabel("Year"); axes[1].set_ylabel("Orders")

        plt.tight_layout()
        plt.show()

# =============================================================================
# STEP 9 — SALES ANALYSIS
# =============================================================================
hdr("STEP 9 — SALES DEEP ANALYSIS", 1)

if "sales" in FRAMES and "products" in FRAMES:
    sf = FRAMES["sales"].copy()
    pf = FRAMES["products"].copy()

    # Detect qty and price columns dynamically
    qty_col   = next((c for c in sf.columns if "quant" in c.lower()), None)
    price_col = next((c for c in sf.columns if "price" in c.lower() or "unit" in c.lower()), None)
    pk_sf     = next((c for c in sf.columns if "product" in c.lower() and "key" in c.lower()), None)
    pk_pf     = next((c for c in pf.columns if "product" in c.lower() and "key" in c.lower()), None)

    if qty_col and price_col:
        sf[qty_col]   = pd.to_numeric(sf[qty_col],   errors="coerce")
        sf[price_col] = pd.to_numeric(sf[price_col], errors="coerce")
        sf["Revenue"] = sf[qty_col] * sf[price_col]

        total_rev = sf["Revenue"].sum()
        avg_order = sf.groupby(next((c for c in sf.columns if "order" in c.lower() and "num" in c.lower()), sf.columns[0]))["Revenue"].sum().mean()

        hdr("Key Sales KPIs", 3)
        display(pd.DataFrame([{
            "Total Revenue":      f"${total_rev:,.2f}",
            "Avg Order Value":    f"${avg_order:,.2f}",
            "Total Transactions": f"{len(sf):,}",
            "Unique Products":    sf[pk_sf].nunique() if pk_sf else "—",
        }]))

        # Merge with products for category analysis
        if pk_sf and pk_pf:
            merged = sf.merge(pf[[pk_pf] + [c for c in pf.columns if "categ" in c.lower() or "name" in c.lower()[:4]]],
                              left_on=pk_sf, right_on=pk_pf, how="left")
            cat_col = next((c for c in merged.columns if "subcateg" not in c.lower() and "categ" in c.lower()), None)

            if cat_col:
                cat_rev = merged.groupby(cat_col)["Revenue"].sum().sort_values(ascending=False)
                fig, ax = plt.subplots(figsize=(12, 5))
                cat_rev.plot(kind="bar", ax=ax, color=PALETTE["primary"], edgecolor="white", alpha=0.9)
                ax.set_title("Revenue by Product Category", fontsize=13, fontweight="bold")
                ax.set_ylabel("Revenue ($)")
                ax.set_xlabel("Category")
                ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f"${x/1e6:.1f}M"))
                plt.xticks(rotation=30, ha="right")
                plt.tight_layout()
                plt.show()

# =============================================================================
# STEP 10 — CUSTOMER ANALYSIS
# =============================================================================
hdr("STEP 10 — CUSTOMER ANALYSIS", 1)

if "customers" in FRAMES:
    cf = FRAMES["customers"].copy()

    # Geography
    country_col = next((c for c in cf.columns if "country" in c.lower()), None)
    gender_col  = next((c for c in cf.columns if "gender" in c.lower()), None)
    bday_col    = next((c for c in cf.columns if "birth" in c.lower() or "dob" in c.lower()), None)

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle("Customer Demographics", fontsize=14, fontweight="bold")

    if country_col:
        top_countries = cf[country_col].value_counts().head(10)
        axes[0].barh(top_countries.index, top_countries.values,
                     color=PALETTE["primary"], alpha=0.85)
        axes[0].set_title("Top 10 Countries")
        axes[0].set_xlabel("Customer Count")
        axes[0].invert_yaxis()

    if gender_col:
        gender_counts = cf[gender_col].value_counts()
        axes[1].pie(gender_counts.values, labels=gender_counts.index, autopct="%1.1f%%",
                    colors=[PALETTE["primary"], PALETTE["accent"]])
        axes[1].set_title("Gender Distribution")

    if bday_col:
        bday_parsed = pd.to_datetime(cf[bday_col], errors="coerce")
        cf["Age"] = (pd.Timestamp.now() - bday_parsed).dt.days // 365
        cf["Age"].dropna().hist(ax=axes[2], bins=20,
                                color=PALETTE["secondary"], edgecolor="white", alpha=0.85)
        axes[2].set_title("Age Distribution")
        axes[2].set_xlabel("Age (years)")

    plt.tight_layout()
    plt.show()

    hdr("Customer Summary Stats", 3)
    display(cf.describe(include="all").T.head(20))

# =============================================================================
# STEP 11 — PRODUCT ANALYSIS
# =============================================================================
hdr("STEP 11 — PRODUCT ANALYSIS", 1)

if "products" in FRAMES:
    pf = FRAMES["products"].copy()
    price_col = next((c for c in pf.columns if "price" in c.lower()), None)
    cost_col  = next((c for c in pf.columns if "cost"  in c.lower()), None)
    cat_col   = next((c for c in pf.columns if "categ" in c.lower() and "sub" not in c.lower()), None)

    if price_col:
        pf[price_col] = pd.to_numeric(pf[price_col], errors="coerce")
    if cost_col:
        pf[cost_col]  = pd.to_numeric(pf[cost_col],  errors="coerce")
    if price_col and cost_col:
        pf["Margin%"] = ((pf[price_col] - pf[cost_col]) / pf[price_col] * 100).round(2)

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle("Product Analysis", fontsize=14, fontweight="bold")

    if price_col:
        pf[price_col].dropna().hist(ax=axes[0], bins=30,
                                    color=PALETTE["primary"], edgecolor="white", alpha=0.85)
        axes[0].set_title("Unit Price Distribution")
        axes[0].set_xlabel("Price ($)")

    if cat_col:
        cat_counts = pf[cat_col].value_counts()
        cat_counts.plot(kind="bar", ax=axes[1],
                        color=PALETTE["secondary"], edgecolor="white", alpha=0.85)
        axes[1].set_title("Products by Category")
        axes[1].set_ylabel("Count")
        plt.setp(axes[1].get_xticklabels(), rotation=30, ha="right")

    if "Margin%" in pf.columns and cat_col:
        margin_by_cat = pf.groupby(cat_col)["Margin%"].mean().sort_values(ascending=False)
        margin_by_cat.plot(kind="bar", ax=axes[2],
                           color=PALETTE["accent"], edgecolor="white", alpha=0.85)
        axes[2].set_title("Avg Margin% by Category")
        axes[2].set_ylabel("Margin %")
        plt.setp(axes[2].get_xticklabels(), rotation=30, ha="right")

    plt.tight_layout()
    plt.show()

# =============================================================================
# STEP 12 — STORE ANALYSIS
# =============================================================================
hdr("STEP 12 — STORE ANALYSIS", 1)

if "stores" in FRAMES:
    stf = FRAMES["stores"].copy()
    hdr("Store Summary Stats", 3)
    display(stf.describe(include="all").T)

    country_col = next((c for c in stf.columns if "country" in c.lower()), None)
    area_col    = next((c for c in stf.columns if "square" in c.lower() or "area" in c.lower() or "sqm" in c.lower()), None)

    if country_col:
        fig, axes = plt.subplots(1, 2, figsize=(16, 5))
        fig.suptitle("Store Analysis", fontsize=14, fontweight="bold")

        store_by_country = stf[country_col].value_counts()
        store_by_country.plot(kind="bar", ax=axes[0],
                              color=PALETTE["primary"], edgecolor="white", alpha=0.85)
        axes[0].set_title("Stores per Country")
        axes[0].set_ylabel("Store Count")
        plt.setp(axes[0].get_xticklabels(), rotation=30, ha="right")

        if area_col:
            stf[area_col] = pd.to_numeric(stf[area_col], errors="coerce")
            stf[area_col].dropna().hist(ax=axes[1], bins=20,
                                        color=PALETTE["secondary"], edgecolor="white", alpha=0.85)
            axes[1].set_title("Store Size Distribution (sqm)")
            axes[1].set_xlabel("Area (sqm)")

        plt.tight_layout()
        plt.show()

# =============================================================================
# STEP 13 — EXCHANGE RATE ANALYSIS
# =============================================================================
hdr("STEP 13 — EXCHANGE RATE ANALYSIS", 1)

if "exchange_rates" in FRAMES:
    ef = FRAMES["exchange_rates"].copy()
    cur_col  = next((c for c in ef.columns if "curr" in c.lower()), None)
    rate_col = next((c for c in ef.columns if "exch" in c.lower() or "rate" in c.lower()), None)
    date_col = next((c for c in ef.columns if "date" in c.lower()), None)

    if cur_col:
        hdr("Currency Coverage", 3)
        display(ef[cur_col].value_counts().reset_index().rename(columns={cur_col: "Currency", "count": "Records"}))

    if cur_col and rate_col and date_col:
        ef[date_col] = pd.to_datetime(ef[date_col], errors="coerce")
        ef[rate_col] = pd.to_numeric(ef[rate_col], errors="coerce")

        top_currencies = ef[cur_col].value_counts().head(6).index.tolist()
        fig, ax = plt.subplots(figsize=(14, 6))
        for cur in top_currencies:
            subset = ef[ef[cur_col] == cur].sort_values(date_col)
            ax.plot(subset[date_col], subset[rate_col], label=cur, linewidth=1.5)
        ax.set_title("Exchange Rate Trends (Top Currencies vs USD)", fontsize=13, fontweight="bold")
        ax.set_xlabel("Date"); ax.set_ylabel("Rate vs USD")
        ax.legend(loc="upper left")
        plt.tight_layout()
        plt.show()

    display(Markdown("""
**Base Currency Normalization Strategy:**
- Designate **USD** as the base currency for all revenue reporting.
- In the dbt intermediate layer: `revenue_usd = quantity * unit_price / exchange_rate`
  where `exchange_rate` is joined on `(order_date, currency_code)`.
- Store both original-currency and USD-normalized revenue in FACT_SALES.
- Alert on missing exchange rate records (orphaned currency/date combos).
"""))

# =============================================================================
# STEP 14 — FEATURE ENGINEERING IDEAS
# =============================================================================
hdr("STEP 14 — FEATURE ENGINEERING IDEAS", 1)

FEATURES = [
    ("revenue_usd",            "quantity * unit_price / exchange_rate",         "Normalize multi-currency revenue to USD for apples-to-apples comparison."),
    ("profit_usd",             "revenue_usd - (quantity * unit_cost_usd)",       "Gross profit per line item; critical for margin dashboards."),
    ("profit_margin_pct",      "profit_usd / revenue_usd * 100",                 "Percentage margin per transaction; highlights premium vs commodity products."),
    ("delivery_days",          "delivery_date - order_date",                     "Fulfilment speed KPI; flag outliers (negative or >90 days)."),
    ("customer_age",           "today - customer_birthdate (in years)",          "Enables age-based segmentation in Power BI."),
    ("customer_lifetime_value","SUM(revenue_usd) per customer",                  "Total spend per customer; foundational for RFM and retention modelling."),
    ("order_size",             "SUM(quantity) per order_number",                 "Basket size metric; correlates with promotional effectiveness."),
    ("store_revenue_sqm",      "store_annual_revenue / store_area_sqm",          "Efficiency KPI for store operations benchmarking."),
    ("days_since_last_order",  "today - MAX(order_date) per customer",           "Recency component of RFM; drives churn-risk segmentation."),
    ("rolling_30d_revenue",    "rolling 30-day SUM of revenue_usd per store",    "Smooth short-term trend line; useful for anomaly detection dashboards."),
    ("yoy_growth_pct",         "(this_year_rev - last_year_rev) / last_year_rev","Year-over-year comparison; core executive KPI."),
    ("category_revenue_share", "category_revenue / total_revenue * 100",         "Portfolio mix analysis; spot over-reliance on single categories."),
]

feat_df = pd.DataFrame(FEATURES, columns=["Feature", "Formula/Logic", "Why Useful"])
display(feat_df.style.set_caption("Recommended Engineered Features").set_properties(**{"text-align": "left"}))

# =============================================================================
# STEP 15 — DBT MODELLING ARCHITECTURE
# =============================================================================
hdr("STEP 15 — dbt MODELLING ARCHITECTURE", 1)

DBT_PLAN = """
## 📦 Staging Layer (`models/staging/`)
| Model | Source | Purpose |
|-------|--------|---------|
| `stg_sales` | Sales.csv | Cast types, rename cols, parse dates, trim strings |
| `stg_customers` | Customers.csv | Standardize name/address, parse DOB, derive age |
| `stg_products` | Products.csv | Cast prices/costs, normalize category names |
| `stg_stores` | Stores.csv | Cast area, parse open_date, normalize country |
| `stg_exchange_rates` | Exchange_Rates.csv | Cast rate, parse date, deduplicate on (date, currency) |

## 🔗 Intermediate Layer (`models/intermediate/`)
| Model | Depends On | Purpose |
|-------|-----------|---------|
| `int_sales_enriched` | stg_sales + stg_exchange_rates | Join exchange rates, compute revenue_usd |
| `int_sales_with_profit` | int_sales_enriched + stg_products | Add unit_cost, compute profit_usd, margin% |
| `int_customer_orders` | int_sales_with_profit + stg_customers | Aggregate CLV, order count, recency per customer |
| `int_store_metrics` | int_sales_with_profit + stg_stores | Aggregate revenue, orders, revenue/sqm per store |

## 🏪 Mart Layer (`models/marts/`)
| Model | Purpose | Power BI Usage |
|-------|---------|----------------|
| `mart_sales_fact` | Final enriched fact table (USD normalized) | Core fact table for all dashboards |
| `mart_dim_customer` | Customer master with CLV + segment | Customer analytics page |
| `mart_dim_product` | Product master with margin + category | Product performance page |
| `mart_dim_store` | Store master with metrics | Store performance page |
| `mart_dim_date` | Full date spine (day/week/month/quarter/year) | Time intelligence in Power BI |
| `mart_executive_summary` | Pre-aggregated monthly KPIs | Executive dashboard (fast load) |

## 🧪 Data Tests to Add
- `not_null` on all PK columns
- `unique` on all PK columns
- `relationships` for all FK columns
- `accepted_values` for gender, currency_code
- `dbt_utils.expression_is_true` for `revenue_usd >= 0`
"""
display(Markdown(DBT_PLAN))

# =============================================================================
# STEP 16 — POWER BI DASHBOARD PLANNING
# =============================================================================
hdr("STEP 16 — POWER BI DASHBOARD PLANNING", 1)

PBI_PLAN = """
### Page 1 — 🏆 Executive Dashboard
**Business Purpose:** C-suite overview of revenue, growth, and portfolio health.
| Visual | Measure | Filter |
|--------|---------|--------|
| KPI Cards | Total Revenue USD, YoY Growth%, Avg Order Value, Total Orders | Year, Region |
| Line Chart | Monthly Revenue Trend (with prior-year overlay) | Country, Category |
| Bar Chart | Revenue by Category | Year |
| Map | Revenue by Country (bubble map) | Year |
| Gauge | Revenue vs Budget (if available) | — |

### Page 2 — 🛒 Sales Analytics
**Business Purpose:** Deep dive into transaction patterns and time intelligence.
| Visual | Measure | Filter |
|--------|---------|--------|
| Matrix | Revenue by Month × Category | Year |
| Waterfall | MoM Revenue Change | Category |
| Scatter | Avg Order Value vs Order Volume per Store | Country |
| Table | Top 20 Orders by Revenue | — |

### Page 3 — 👤 Customer Analytics
**Business Purpose:** Customer segmentation, loyalty, and lifetime value.
| Visual | Measure | Filter |
|--------|---------|--------|
| Bar Chart | Revenue by Customer Segment | Country, Year |
| Scatter | RFM Segmentation plot | Segment |
| Map | Customers by City | Country |
| KPI | Avg CLV, Repeat Purchase Rate, Avg Recency Days | Year |
| Table | Top 50 Customers | Country |

### Page 4 — 📦 Product Performance
**Business Purpose:** Identify best and worst products by revenue and margin.
| Visual | Measure | Filter |
|--------|---------|--------|
| Bar Chart | Top 20 Products by Revenue | Category |
| Scatter | Revenue vs Margin% per Product | Category |
| Treemap | Revenue share by Category > Subcategory | Year |
| Table | All products with Revenue, Margin%, Rank | Category |

### Page 5 — 🏪 Store Performance
**Business Purpose:** Compare store productivity across geographies.
| Visual | Measure | Filter |
|--------|---------|--------|
| Bar Chart | Revenue per Store | Country |
| Map | Store locations + revenue (bubble) | Country |
| KPI | Revenue per sqm, Avg Store Age | Country |
| Matrix | Store × Month revenue heatmap | Country |

### Page 6 — 💹 Currency & FX Impact
**Business Purpose:** Understand multi-currency exposure and FX conversion effects.
| Visual | Measure | Filter |
|--------|---------|--------|
| Line Chart | Exchange Rate Trends per Currency | Currency |
| Bar Chart | Revenue by Currency (local vs USD) | Year |
| KPI | FX Impact on Revenue (original vs USD) | Year, Currency |
"""
display(Markdown(PBI_PLAN))

# =============================================================================
# STEP 17 — KPI SUGGESTIONS
# =============================================================================
hdr("STEP 17 — KPI SUGGESTION CATALOGUE", 1)

KPIS = [
    ("Total Revenue USD",         "SUM(revenue_usd)",                      "Primary top-line metric for all stakeholders"),
    ("Gross Profit",              "SUM(profit_usd)",                       "Profitability after COGS; essential for margin management"),
    ("Gross Margin %",            "SUM(profit_usd)/SUM(revenue_usd)*100",  "Efficiency of revenue; benchmark across categories/stores"),
    ("Average Order Value (AOV)", "Total Revenue / Distinct Orders",       "Health of basket size; impacted by promotions"),
    ("Units Sold",                "SUM(quantity)",                         "Volume indicator; separates price from volume effects"),
    ("Customer Count",            "DISTINCTCOUNT(customer_key)",           "Market penetration; compare active vs total customers"),
    ("Revenue per Customer",      "Total Revenue / Customer Count",        "Monetization efficiency"),
    ("Repeat Purchase Rate",      "Customers with >1 order / Total",       "Loyalty & retention indicator"),
    ("YoY Revenue Growth %",      "(Current Year - Prior Year)/Prior Year","Executive growth narrative"),
    ("Store Revenue per sqm",     "Store Revenue / Store Area",            "Operational efficiency benchmark"),
    ("CLV (Avg)",                 "AVG(SUM revenue per customer)",         "Future investment signal; marketing ROI denominator"),
    ("Top Category Revenue Share","Category Rev / Total Rev",              "Portfolio concentration risk"),
    ("Delivery Days (Avg)",       "AVG(delivery_date - order_date)",       "Supply chain KPI; customer satisfaction proxy"),
]

kpi_df = pd.DataFrame(KPIS, columns=["KPI Name", "Formula", "Business Meaning"])
display(kpi_df.style.set_caption("Recommended KPI Catalogue").set_properties(**{"text-align": "left"}))

# =============================================================================
# STEP 18 — DAX MEASURE IDEAS
# =============================================================================
hdr("STEP 18 — DAX MEASURE IDEAS", 1)

DAX_MEASURES = """
```dax
-- 1. Total Revenue
Total Revenue = SUM(mart_sales_fact[revenue_usd])

-- 2. Prior Year Revenue
PY Revenue = CALCULATE([Total Revenue], SAMEPERIODLASTYEAR(mart_dim_date[Date]))

-- 3. YoY Growth %
YoY Growth % = DIVIDE([Total Revenue] - [PY Revenue], [PY Revenue], 0)

-- 4. Running Total Revenue (time-ordered)
Running Total Revenue =
    CALCULATE([Total Revenue],
              FILTER(ALL(mart_dim_date), mart_dim_date[Date] <= MAX(mart_dim_date[Date])))

-- 5. Average Order Value
Avg Order Value = DIVIDE([Total Revenue], DISTINCTCOUNT(mart_sales_fact[order_number]))

-- 6. Gross Profit
Gross Profit = SUM(mart_sales_fact[profit_usd])

-- 7. Gross Margin %
Gross Margin % = DIVIDE([Gross Profit], [Total Revenue], 0)

-- 8. Revenue per Customer
Revenue per Customer = DIVIDE([Total Revenue], DISTINCTCOUNT(mart_sales_fact[customer_key]))

-- 9. MoM Revenue Growth
MoM Growth % =
VAR currentMonth = [Total Revenue]
VAR priorMonth   = CALCULATE([Total Revenue], DATEADD(mart_dim_date[Date], -1, MONTH))
RETURN DIVIDE(currentMonth - priorMonth, priorMonth, 0)

-- 10. Top N Products (parameter-driven)
Top N Revenue =
    CALCULATE([Total Revenue],
              TOPN([N Parameter], ALL(mart_dim_product[product_name]), [Total Revenue]))
```
"""
display(Markdown(DAX_MEASURES))

# =============================================================================
# STEP 19 — FINAL SUMMARY REPORT
# =============================================================================
hdr("STEP 19 — FINAL PROFESSIONAL SUMMARY REPORT", 1)

SUMMARY = f"""
## Dataset Overview
- **6 tables** loaded covering: Customers, Products, Stores, Sales transactions, Exchange Rates, and a Data Dictionary.
- **Sales** is the central fact table. All other tables are dimension/reference tables.
- Dataset spans multiple countries and currencies — **USD normalization is mandatory** before reporting.

## Business Domain
Global multi-store **retail business** with:
- Customers across multiple continents
- Products organised into categories and subcategories
- Physical store network with varying sizes
- Sales recorded in local currencies → FX conversion required

## Relationship Architecture
```
SALES (FACT)
  ├── → CUSTOMERS   (many-to-one on CustomerKey)
  ├── → PRODUCTS    (many-to-one on ProductKey)
  ├── → STORES      (many-to-one on StoreKey)
  └── → EXCHANGE_RATES (many-to-one on Date + CurrencyCode)
```

## Data Quality Findings
See STEP 7 for detailed severity ranking.
Key risks:
1. Multi-currency revenue — **not comparable without FX conversion**
2. Potential orphan records in Sales FK joins — verify all keys match dimensions
3. Date column parsing issues possible — validate OrderDate and DeliveryDate ranges

## Recommended Cleaning Steps (in dbt Staging)
1. Trim whitespace on all string columns
2. Standardize casing (UPPER country codes, Title Case names)
3. Cast all date columns to DATE type
4. Cast price/cost/rate columns to FLOAT
5. Deduplicate exchange_rates on (date, currency)
6. Validate FK integrity and log orphan records

## Recommended dbt Structure
`staging → intermediate → marts`
See STEP 15 for full model list.

## Recommended Dashboard Strategy
6 Power BI pages: Executive, Sales, Customers, Products, Stores, Currency.
All pages share slicers for Year and Country.
See STEP 16 for detailed layout.

## Major Business Opportunities
- **RFM Segmentation**: Identify high-value vs at-risk customers
- **Margin Optimization**: Category and product profitability ranking
- **Store Efficiency**: Revenue per sqm league table
- **Seasonal Promotions**: Monthly trend + YoY gap analysis
- **Currency Hedging Insight**: Revenue exposure by currency

## Risks & Challenges
- Missing exchange rates for some date-currency combos
- No explicit budget/target data → limits variance reporting
- No returns/refund data → reported revenue may be gross-only
- Store open dates needed for cohort-adjusted performance

## Suggested Next Steps
1. ✅ Run this notebook and review all outputs
2. 🔧 Fix data quality issues flagged in STEP 7
3. 🏗️ Initialise dbt project: `dbt init retail_bi`
4. 📄 Build staging models (one per source table)
5. 🔗 Build intermediate models (sales_enriched, store_metrics)
6. 🏪 Build mart models (fact + dims + dim_date spine)
7. 📊 Connect Power BI to dbt mart layer via connector
8. 📈 Build dashboards page by page using STEP 16 blueprint
"""
display(Markdown(SUMMARY))

# =============================================================================
# STEP 20 — BONUS: CORRELATION, OUTLIER & RFM ANALYSIS
# =============================================================================
hdr("STEP 20 — BONUS ADVANCED ANALYSIS", 1)

# Correlation heatmap (numeric columns across sales + products)
hdr("Correlation Analysis — Sales & Products", 3)
if "sales" in FRAMES and "products" in FRAMES:
    sf = FRAMES["sales"].copy()
    pf = FRAMES["products"].copy()
    pk_sf = next((c for c in sf.columns if "product" in c.lower() and "key" in c.lower()), None)
    pk_pf = next((c for c in pf.columns if "product" in c.lower() and "key" in c.lower()), None)
    if pk_sf and pk_pf:
        merged = sf.merge(pf, left_on=pk_sf, right_on=pk_pf, how="left", suffixes=("_sale","_prod"))
        num_cols = merged.select_dtypes(include=[np.number]).columns[:12]
        if len(num_cols) > 2:
            corr = merged[num_cols].corr()
            fig, ax = plt.subplots(figsize=(12, 8))
            sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm",
                        center=0, ax=ax, linewidths=0.5)
            ax.set_title("Correlation Matrix — Sales × Products", fontsize=13, fontweight="bold")
            plt.tight_layout()
            plt.show()

# RFM Framework
hdr("RFM Segmentation Framework", 3)
RFM_TEXT = """
**RFM** (Recency, Frequency, Monetary) segmentation maps every customer to a 3-digit score
(1–5 per dimension) enabling targeted marketing:

| Segment | R | F | M | Strategy |
|---------|---|---|---|----------|
| Champions | 5 | 5 | 5 | Reward, upsell |
| Loyal | 4–5 | 4–5 | 3–5 | Loyalty programme |
| At Risk | 2–3 | 3–4 | 3–4 | Win-back campaign |
| Hibernating | 1–2 | 1–2 | 1–2 | Reactivation offer |
| Lost | 1 | 1 | 1 | Last-chance email |

**To compute in dbt `int_customer_rfm`:**
```sql
SELECT
    customer_key,
    DATEDIFF('day', MAX(order_date), CURRENT_DATE)           AS recency_days,
    COUNT(DISTINCT order_number)                             AS frequency,
    SUM(revenue_usd)                                         AS monetary_usd,
    NTILE(5) OVER (ORDER BY recency_days DESC)               AS r_score,
    NTILE(5) OVER (ORDER BY frequency)                       AS f_score,
    NTILE(5) OVER (ORDER BY monetary_usd)                    AS m_score
FROM mart_sales_fact
GROUP BY 1
```
"""
display(Markdown(RFM_TEXT))

# Outlier detection on revenue
if "sales" in FRAMES:
    sf = FRAMES["sales"].copy()
    qty_col   = next((c for c in sf.columns if "quant" in c.lower()), None)
    price_col = next((c for c in sf.columns if "price" in c.lower() or "unit" in c.lower()), None)
    if qty_col and price_col:
        sf[qty_col]   = pd.to_numeric(sf[qty_col],   errors="coerce")
        sf[price_col] = pd.to_numeric(sf[price_col], errors="coerce")
        sf["Revenue"] = sf[qty_col] * sf[price_col]
        Q1 = sf["Revenue"].quantile(0.25)
        Q3 = sf["Revenue"].quantile(0.75)
        IQR = Q3 - Q1
        outliers = sf[(sf["Revenue"] < Q1 - 1.5*IQR) | (sf["Revenue"] > Q3 + 1.5*IQR)]
        hdr(f"Revenue Outliers Detected: {len(outliers):,} rows ({len(outliers)/len(sf)*100:.1f}% of sales)", 3)
        display(outliers.head(10))

        fig, ax = plt.subplots(figsize=(12, 4))
        ax.boxplot(sf["Revenue"].dropna(), vert=False, patch_artist=True,
                   boxprops=dict(facecolor=PALETTE["primary"], alpha=0.6))
        ax.set_title("Revenue Distribution with Outlier Bounds (IQR method)", fontsize=12, fontweight="bold")
        ax.set_xlabel("Revenue ($)")
        plt.tight_layout()
        plt.show()

hdr("✅ ANALYSIS COMPLETE — Ready for dbt Modelling & Power BI", 1)
display(Markdown("""
**Next Actions:**
1. Review all flagged data quality issues above.
2. Use the dbt architecture from STEP 15 to scaffold your models.
3. Use the Power BI blueprint from STEP 16 to start dashboard design.
4. Implement RFM in `int_customer_rfm` dbt model.
5. Build the DAX measures from STEP 18 in Power BI.
"""))
````

````python
# =============================================================================
# RETAIL DATASET — COMPLETE DEEP ANALYSIS (FINAL CONSOLIDATED VERSION)
# Steps 1–20 + YoY + Top Products — Run entirely in Google Colab
# =============================================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
import os
from IPython.display import display, Markdown

warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 50)
pd.set_option("display.max_colwidth", 80)
pd.set_option("display.float_format", lambda x: f"{x:,.4f}")

PALETTE = {
    "primary":   "#2563EB",
    "secondary": "#7C3AED",
    "accent":    "#F59E0B",
    "danger":    "#EF4444",
    "success":   "#10B981",
    "neutral":   "#6B7280",
}

sns.set_theme(style="whitegrid", palette="muted", font_scale=1.05)

def hdr(text, level=2):
    display(Markdown("#" * level + " " + text))

def info(text):
    display(Markdown(f"> ℹ️  {text}"))

def warn(text):
    display(Markdown(f"> ⚠️  **WARNING:** {text}"))

def ok(text):
    display(Markdown(f"> ✅  {text}"))

def fmt_usd(x):
    return mticker.FuncFormatter(lambda v, _: f"${v/x:.1f}{'M' if x==1e6 else 'K'}")

# =============================================================================
# STEP 1 — LOAD ALL DATASETS
# =============================================================================
hdr("STEP 1 — LOAD ALL DATASETS", 1)
info("Loading all 6 CSV files with automatic encoding & delimiter detection.")

FILE_PATHS = {
    "customers":      "Customers.csv",
    "data_dict":      "Data_Dictionary.csv",
    "exchange_rates": "Exchange_Rates.csv",
    "products":       "Products.csv",
    "sales":          "Sales.csv",
    "stores":         "Stores.csv",
}

def smart_load(path):
    for enc in ["utf-8", "latin-1", "cp1252"]:
        try:
            df = pd.read_csv(path, encoding=enc, sep=None, engine="python",
                             on_bad_lines="warn")
            return df, enc
        except Exception:
            continue
    raise ValueError(f"Cannot read {path}")

FRAMES = {}
LOAD_META = []
for name, path in FILE_PATHS.items():
    if not os.path.exists(path):
        warn(f"`{path}` not found — skipping.")
        continue
    df, enc = smart_load(path)
    FRAMES[name] = df
    LOAD_META.append({
        "Table":       name,
        "File":        path,
        "Rows":        len(df),
        "Columns":     len(df.columns),
        "Encoding":    enc,
        "Memory (KB)": round(df.memory_usage(deep=True).sum() / 1024, 1),
    })

display(pd.DataFrame(LOAD_META).style
        .set_caption("All Loaded Tables")
        .bar(subset=["Rows"],       color=PALETTE["primary"])
        .bar(subset=["Memory (KB)"],color=PALETTE["accent"]))

# =============================================================================
# STEP 1b — CLEAN KNOWN ISSUES IMMEDIATELY AFTER LOAD
# Fixes: $ signs + trailing spaces in Products price columns
# =============================================================================
hdr("STEP 1b — APPLY KNOWN FIXES (Price Columns)", 2)

for col in ["Unit Cost USD", "Unit Price USD"]:
    FRAMES["products"][col] = (
        FRAMES["products"][col]
        .astype(str).str.strip()
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .astype(float)
    )
ok("Unit Cost USD and Unit Price USD cleaned — $ signs removed, cast to float.")

# =============================================================================
# STEP 2 — COLUMN-LEVEL ANALYSIS
# =============================================================================
hdr("STEP 2 — COLUMN-LEVEL ANALYSIS", 1)
info("Dtypes, nulls, uniques, min/max, string lengths, suspicious values for every table.")

def col_profile(df, table_name):
    hdr(f"Table: `{table_name.upper()}`", 3)
    rows = []
    for col in df.columns:
        s        = df[col]
        null_cnt = s.isna().sum()
        null_pct = round(null_cnt / len(s) * 100, 2)
        unique   = s.nunique()
        dup_cnt  = max(len(s) - unique - null_cnt, 0)
        try:    mn, mx = s.min(), s.max()
        except: mn, mx = "—", "—"
        if s.dtype == object:
            lens     = s.dropna().astype(str).str.len()
            str_info = f"len {int(lens.min())}–{int(lens.max())} (mean {lens.mean():.1f})"
        else:
            str_info = "—"
        sample = s.dropna().unique()[:3].tolist()
        rows.append({
            "Column":      col,
            "Dtype":       str(s.dtype),
            "Nulls":       null_cnt,
            "Null%":       null_pct,
            "Uniques":     unique,
            "Dups (est)":  dup_cnt,
            "Min":         mn,
            "Max":         mx,
            "String Info": str_info,
            "Sample":      str(sample),
        })
    profile_df = pd.DataFrame(rows)
    display(profile_df.style
            .highlight_between(subset=["Null%"], left=20, right=100, color="#FEE2E2")
            .set_caption(f"Column Profile — {table_name}"))

    issues = []
    for col in df.columns:
        s = df[col]
        if s.dtype == object:
            if s.str.contains(r"^\s+|\s+$", na=False).any():
                issues.append(f"**{col}**: trailing/leading spaces")
            if s.str.lower().nunique() < s.nunique():
                issues.append(f"**{col}**: inconsistent casing")
            if (s == "").sum():
                issues.append(f"**{col}**: {(s=='').sum()} empty strings")
        if pd.api.types.is_numeric_dtype(s):
            neg = (s < 0).sum()
            if neg:
                issues.append(f"**{col}**: {neg} negative values")
    if issues:
        warn(f"Findings in `{table_name}`:\n" + "\n".join(f"- {i}" for i in issues))
    else:
        ok(f"No suspicious values in `{table_name}`.")

for name, df in FRAMES.items():
    col_profile(df, name)

# =============================================================================
# STEP 3 — TABLE BUSINESS INTERPRETATION
# =============================================================================
hdr("STEP 3 — TABLE BUSINESS INTERPRETATION", 1)

TABLE_BUSINESS = {
    "customers":      ("Master buyer dimension — demographics, geography, loyalty segmentation.",
                       "DIM_CUSTOMER. Central to RFM, CLV, churn.",
                       "CustomerKey, Gender, City, Country, Birthday"),
    "data_dict":      ("Metadata catalog describing all columns.",
                       "Documentation layer — drives dbt column naming standards.",
                       "Table, Field, Description"),
    "exchange_rates": ("Currency conversion reference for multi-currency normalisation.",
                       "Helper snapshot table — applied in dbt int layer.",
                       "Date, Currency, Exchange"),
    "products":       ("Product master — pricing, cost, category hierarchy.",
                       "DIM_PRODUCT. Drives margin, category, SKU analyses.",
                       "ProductKey, Category, Subcategory, Unit Price, Unit Cost"),
    "sales":          ("Central fact table — every transaction line item.",
                       "FACT_SALES. Joins all dims. Primary KPI source.",
                       "Order Number, Line Item, CustomerKey, StoreKey, ProductKey, Quantity"),
    "stores":         ("Store master — geography and operational metadata.",
                       "DIM_STORE. Regional and store-level rollups.",
                       "StoreKey, Country, State, Square Meters, Open Date"),
}

for tbl, (purpose, role, keys) in TABLE_BUSINESS.items():
    if tbl not in FRAMES:
        continue
    hdr(f"`{tbl.upper()}`", 3)
    display(Markdown(f"""
| Attribute | Detail |
|-----------|--------|
| **Purpose** | {purpose} |
| **BI Role** | {role} |
| **Key Columns** | {keys} |
| **Rows** | {len(FRAMES[tbl]):,} |
"""))

# =============================================================================
# STEP 4 — PRIMARY KEY DETECTION
# =============================================================================
hdr("STEP 4 — PRIMARY KEY DETECTION", 1)

PK_RESULTS = {}
for name, df in FRAMES.items():
    hdr(f"Table: `{name.upper()}`", 3)
    rows = []
    for col in df.columns:
        subset = df[[col]].dropna()
        total  = len(subset)
        uniq   = subset.drop_duplicates().shape[0]
        pct    = round(uniq / total * 100, 2) if total else 0
        dups   = total - uniq
        rows.append({"Column": col, "Unique%": pct, "Duplicates": dups,
                     "Verdict": "✅ Strong PK" if pct == 100 else
                               ("⚠️ Weak" if pct > 90 else "❌ Not PK")})
    pk_df = pd.DataFrame(rows).sort_values("Unique%", ascending=False)
    display(pk_df.style.set_caption(f"PK Analysis — {name}"))
    best = pk_df[pk_df["Unique%"] == 100]
    if not best.empty:
        ok(f"Likely PK for `{name}`: `{best.iloc[0]['Column']}`")
        PK_RESULTS[name] = best.iloc[0]["Column"]
    else:
        warn(f"No single-column PK for `{name}`. Composite key needed "
             f"(e.g. Order Number + Line Item for sales).")

# =============================================================================
# STEP 5 — FOREIGN KEY & RELATIONSHIP DETECTION
# =============================================================================
hdr("STEP 5 — FOREIGN KEY & RELATIONSHIP DETECTION", 1)

EXPECTED_RELS = [
    ("sales", "CustomerKey", "customers",  "CustomerKey"),
    ("sales", "ProductKey",  "products",   "ProductKey"),
    ("sales", "StoreKey",    "stores",     "StoreKey"),
]

rel_rows = []
for (fact_tbl, fact_col, dim_tbl, dim_col) in EXPECTED_RELS:
    if fact_tbl not in FRAMES or dim_tbl not in FRAMES:
        continue
    fact_df   = FRAMES[fact_tbl]
    dim_df    = FRAMES[dim_tbl]
    fc        = next((c for c in fact_df.columns if fact_col.lower() in c.lower()), None)
    dc        = next((c for c in dim_df.columns  if dim_col.lower()  in c.lower()), None)
    if not fc or not dc:
        warn(f"Column not found: {fact_tbl}.{fact_col}")
        continue
    fact_keys = set(fact_df[fc].dropna().unique())
    dim_keys  = set(dim_df[dc].dropna().unique())
    orphans   = fact_keys - dim_keys
    match_pct = round((len(fact_keys - orphans) / len(fact_keys)) * 100, 2) if fact_keys else 0
    rel_rows.append({
        "Relationship": f"{fact_tbl}.{fc} → {dim_tbl}.{dc}",
        "Fact Keys":    len(fact_keys),
        "Dim Keys":     len(dim_keys),
        "Match%":       match_pct,
        "Orphans":      len(orphans),
        "Cardinality":  "Many-to-One",
        "Status":       "✅ OK" if match_pct == 100 else "⚠️ Orphans exist",
    })

display(pd.DataFrame(rel_rows).style.set_caption("FK Relationship Audit"))

# =============================================================================
# STEP 6 — ER DIAGRAM
# =============================================================================
hdr("STEP 6 — ER DIAGRAM", 1)

print("""
┌─────────────────────┐        ┌──────────────────────────┐
│     CUSTOMERS        │        │          SALES            │
│─────────────────────│        │──────────────────────────│
│ PK: CustomerKey     │◄───────│ FK: CustomerKey           │
│     Gender          │        │ FK: ProductKey            │
│     Name            │        │ FK: StoreKey              │
│     City / State    │        │ PK: (OrderNumber+LineItem)│
│     Country         │        │     OrderDate             │
│     Birthday        │        │     DeliveryDate          │
└─────────────────────┘        │     Quantity              │
                                │     CurrencyCode          │
┌─────────────────────┐        └──────────────────────────┘
│     PRODUCTS         │◄──────────────┘
│─────────────────────│
│ PK: ProductKey      │        ┌──────────────────────────┐
│     Product Name    │        │     EXCHANGE_RATES        │
│     Brand / Color   │        │──────────────────────────│
│     Category        │        │ PK: (Date + Currency)    │
│     Unit Price USD  │        │     Exchange Rate         │
│     Unit Cost USD   │        └──────────────────────────┘
└─────────────────────┘
                                ┌──────────────────────────┐
┌─────────────────────┐        │     DATA_DICTIONARY       │
│     STORES           │◄───── │──────────────────────────│
│─────────────────────│        │     Table / Field         │
│ PK: StoreKey        │        │     Description           │
│     Country / State │        └──────────────────────────┘
│     Square Meters   │
│     Open Date       │
└─────────────────────┘
""")

display(Markdown("""
```mermaid
erDiagram
    CUSTOMERS { int CustomerKey PK; string Gender; string Name; string Country; date Birthday }
    PRODUCTS  { int ProductKey PK; string ProductName; string Category; float UnitCostUSD; float UnitPriceUSD }
    STORES    { int StoreKey PK; string Country; string State; int SquareMeters; date OpenDate }
    EXCHANGE_RATES { date Date; string Currency; float Exchange }
    SALES { int OrderNumber PK; int LineItem PK; date OrderDate; int CustomerKey FK; int StoreKey FK; int ProductKey FK; int Quantity; string CurrencyCode }

    CUSTOMERS      ||--o{ SALES : "places"
    PRODUCTS       ||--o{ SALES : "appears in"
    STORES         ||--o{ SALES : "fulfilled by"
    EXCHANGE_RATES }o--|| SALES : "currency ref"
```
"""))

# =============================================================================
# STEP 7 — DATA QUALITY AUDIT
# =============================================================================
hdr("STEP 7 — DATA QUALITY AUDIT", 1)

quality_issues = {"HIGH": [], "MEDIUM": [], "LOW": []}

for name, df in FRAMES.items():
    dup_rows = df.duplicated().sum()
    if dup_rows:
        quality_issues["HIGH"].append(f"`{name}`: {dup_rows} fully duplicate rows")
    for col in df.columns:
        pct = df[col].isna().mean() * 100
        if pct > 50:   quality_issues["HIGH"].append(f"`{name}.{col}`: {pct:.1f}% nulls")
        elif pct > 20: quality_issues["MEDIUM"].append(f"`{name}.{col}`: {pct:.1f}% nulls")
        elif pct > 5:  quality_issues["LOW"].append(f"`{name}.{col}`: {pct:.1f}% nulls")
    for col in df.select_dtypes(include=[np.number]).columns:
        neg = (df[col] < 0).sum()
        if neg:
            sev = "HIGH" if any(k in col.lower() for k in ["price","qty","quantity","rev"]) else "MEDIUM"
            quality_issues[sev].append(f"`{name}.{col}`: {neg} negative values")
    for col in df.columns:
        if "date" in col.lower():
            try:
                parsed  = pd.to_datetime(df[col], errors="coerce")
                future  = (parsed > pd.Timestamp.now()).sum()
                if future: quality_issues["MEDIUM"].append(f"`{name}.{col}`: {future} future dates")
            except: pass

for level, items in quality_issues.items():
    icon = {"HIGH":"🔴","MEDIUM":"🟡","LOW":"🟢"}[level]
    hdr(f"{icon} {level} RISK ISSUES ({len(items)})", 3)
    for item in items:
        display(Markdown(f"- {item}"))
    if not items:
        ok("No issues at this level.")

display(Markdown("""
**Key Finding:** `sales.Delivery Date` is 79.1% null.
This is **by design** — StoreKey=0 represents the online channel where delivery date tracking
was not captured. Split analyses into Online (StoreKey=0) vs In-Store (StoreKey>0).
"""))

# =============================================================================
# STEP 8 — TIME ANALYSIS
# =============================================================================
hdr("STEP 8 — TIME ANALYSIS", 1)

for name, df in FRAMES.items():
    date_cols = [c for c in df.columns if "date" in c.lower()]
    if not date_cols:
        continue
    hdr(f"`{name.upper()}` — Date Columns", 3)
    for col in date_cols:
        parsed = pd.to_datetime(df[col], errors="coerce")
        valid  = parsed.dropna()
        if valid.empty:
            warn(f"`{col}` — no parseable dates.")
            continue
        info(f"`{col}` | Range: {valid.min().date()} → {valid.max().date()} "
             f"| Nulls: {parsed.isna().sum():,} | Valid: {len(valid):,}")

# =============================================================================
# STEP 9 — SALES DEEP ANALYSIS (COMPLETE WITH YoY + TOP PRODUCTS)
# =============================================================================
hdr("STEP 9 — SALES DEEP ANALYSIS", 1)

sf = FRAMES["sales"].copy()
pf = FRAMES["products"].copy()

# Parse dates
sf["Order Date"]    = pd.to_datetime(sf["Order Date"],    errors="coerce")
sf["Delivery Date"] = pd.to_datetime(sf["Delivery Date"], errors="coerce")

# Merge ALL product columns so nothing goes missing
merged = sf.merge(pf, on="ProductKey", how="left")

# Computed columns
merged["Revenue"]    = merged["Quantity"] * merged["Unit Price USD"]
merged["Cost"]       = merged["Quantity"] * merged["Unit Cost USD"]
merged["Profit"]     = merged["Revenue"]  - merged["Cost"]
merged["Margin_pct"] = (merged["Profit"]  / merged["Revenue"] * 100).round(2)
merged["Year"]       = merged["Order Date"].dt.year
merged["Month"]      = merged["Order Date"].dt.month
merged["YearMonth"]  = merged["Order Date"].dt.to_period("M")

# ── KPI SUMMARY ──────────────────────────────────────────────────────────────
total_rev    = merged["Revenue"].sum()
total_profit = merged["Profit"].sum()
avg_margin   = merged["Margin_pct"].mean()
n_orders     = merged["Order Number"].nunique()
aov          = total_rev / n_orders
total_units  = merged["Quantity"].sum()

print("=" * 57)
print("          SALES KPI SUMMARY (Pre-FX, USD List Price)")
print("=" * 57)
print(f"  Total Revenue       : ${total_rev:>15,.2f}")
print(f"  Total Gross Profit  : ${total_profit:>15,.2f}")
print(f"  Avg Gross Margin %  : {avg_margin:>15.2f}%")
print(f"  Total Orders        : {n_orders:>15,}")
print(f"  Avg Order Value     : ${aov:>15,.2f}")
print(f"  Total Line Items    : {len(merged):>15,}")
print(f"  Total Units Sold    : {total_units:>15,}")
print("=" * 57)

# ── CHART BLOCK 1: Monthly trend + Category + Annual + Top 10 products ───────
monthly   = merged.groupby("YearMonth").agg(Revenue=("Revenue","sum")).reset_index()
monthly["YearMonth"] = monthly["YearMonth"].astype(str)

cat_rev   = merged.groupby("Category")["Revenue"].sum().sort_values(ascending=False)
yearly    = merged.groupby("Year")["Revenue"].sum()
top_prods = (merged.groupby(["ProductKey","Product Name","Category"])["Revenue"]
             .sum().nlargest(10).reset_index())

fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle("STEP 9 — Sales Deep Analysis", fontsize=16, fontweight="bold", y=1.01)

# Monthly trend
axes[0,0].fill_between(range(len(monthly)), monthly["Revenue"],
                        alpha=0.3, color=PALETTE["primary"])
axes[0,0].plot(range(len(monthly)), monthly["Revenue"],
               color=PALETTE["primary"], linewidth=2)
axes[0,0].set_xticks(range(0, len(monthly), 6))
axes[0,0].set_xticklabels(monthly["YearMonth"].iloc[::6], rotation=45, ha="right")
axes[0,0].set_title("Monthly Revenue Trend", fontweight="bold")
axes[0,0].set_ylabel("Revenue ($)")
axes[0,0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f"${x/1e6:.1f}M"))

# Category bar
axes[0,1].bar(cat_rev.index, cat_rev.values,
              color=PALETTE["secondary"], edgecolor="white", alpha=0.9)
axes[0,1].set_title("Revenue by Product Category", fontweight="bold")
axes[0,1].set_ylabel("Revenue ($)")
axes[0,1].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f"${x/1e6:.1f}M"))
plt.setp(axes[0,1].get_xticklabels(), rotation=30, ha="right")

# Annual bar
axes[1,0].bar(yearly.index.astype(str), yearly.values,
              color=PALETTE["accent"], edgecolor="white", alpha=0.9)
axes[1,0].set_title("Annual Revenue", fontweight="bold")
axes[1,0].set_ylabel("Revenue ($)")
axes[1,0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f"${x/1e6:.1f}M"))

# Top 10 products (named)
labels = [f"{row['Product Name'][:35]}..." if len(row['Product Name'])>35
          else row['Product Name'] for _, row in top_prods.iterrows()]
axes[1,1].barh(labels, top_prods["Revenue"].values,
               color=PALETTE["success"], edgecolor="white", alpha=0.9)
axes[1,1].set_title("Top 10 Products by Revenue", fontweight="bold")
axes[1,1].set_xlabel("Revenue ($)")
axes[1,1].invert_yaxis()
axes[1,1].xaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f"${x/1e3:.0f}K"))

plt.tight_layout()
plt.show()

# ── YoY GROWTH TABLE ─────────────────────────────────────────────────────────
hdr("Year-over-Year Revenue Growth", 3)
yoy = merged.groupby("Year")["Revenue"].sum().reset_index()
yoy["YoY_Growth%"] = yoy["Revenue"].pct_change() * 100
yoy["Revenue_fmt"] = yoy["Revenue"].map("${:,.0f}".format)
yoy["YoY_Growth%"] = yoy["YoY_Growth%"].map(lambda x: f"{x:+.1f}%" if pd.notna(x) else "—")
display(yoy[["Year","Revenue_fmt","YoY_Growth%"]].rename(
    columns={"Revenue_fmt":"Revenue","YoY_Growth%":"YoY Growth %"}))
warn("2021 covers Jan–Feb only — exclude from YoY comparisons in Power BI.")

# ── PARETO CHECK ──────────────────────────────────────────────────────────────
hdr("Pareto / ABC Analysis", 3)
prod_rev   = merged.groupby("ProductKey")["Revenue"].sum().sort_values(ascending=False)
cumulative = prod_rev.cumsum() / prod_rev.sum() * 100
top20_pct  = cumulative.iloc[int(len(cumulative) * 0.2)]
print(f"Top 20% of SKUs ({int(len(prod_rev)*0.2)} products) drive {top20_pct:.1f}% of revenue.")
print("→ Build ABC classification (A=top 70%, B=next 20%, C=bottom 10%) in dbt mart layer.")

# ── TOP 10 PRODUCTS TABLE ─────────────────────────────────────────────────────
hdr("Top 10 Products — Full Detail", 3)
top10_display = top_prods.copy()
top10_display["Revenue"] = top10_display["Revenue"].map("${:,.0f}".format)
display(top10_display)

# ── CATEGORY BREAKDOWN TABLE ──────────────────────────────────────────────────
hdr("Category Revenue Breakdown", 3)
cat_summary = merged.groupby("Category").agg(
    Revenue    =("Revenue",    "sum"),
    Profit     =("Profit",     "sum"),
    Orders     =("Order Number","nunique"),
    Units      =("Quantity",   "sum"),
    Avg_Margin =("Margin_pct", "mean"),
).sort_values("Revenue", ascending=False).round(2)
cat_summary["Revenue_Share%"] = (cat_summary["Revenue"] / total_rev * 100).round(1)
cat_summary["Revenue"] = cat_summary["Revenue"].map("${:,.0f}".format)
cat_summary["Profit"]  = cat_summary["Profit"].map("${:,.0f}".format)
display(cat_summary)

# ── SEASONALITY HEATMAP ───────────────────────────────────────────────────────
hdr("Monthly Seasonality Heatmap", 3)
pivot = merged.pivot_table(values="Revenue", index="Year", columns="Month",
                            aggfunc="sum") / 1e6
fig, ax = plt.subplots(figsize=(14, 5))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap="Blues", ax=ax,
            linewidths=0.5, cbar_kws={"label": "Revenue ($M)"})
ax.set_title("Monthly Revenue Heatmap by Year ($M)", fontweight="bold")
ax.set_xlabel("Month"); ax.set_ylabel("Year")
plt.tight_layout()
plt.show()

# =============================================================================
# STEP 10 — CUSTOMER ANALYSIS
# =============================================================================
hdr("STEP 10 — CUSTOMER ANALYSIS", 1)

cf = FRAMES["customers"].copy()
cf["Birthday"] = pd.to_datetime(cf["Birthday"], errors="coerce")
cf["Age"]      = ((pd.Timestamp.now() - cf["Birthday"]).dt.days / 365).round(0)

country_col = "Country"
gender_col  = "Gender"

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Customer Demographics", fontsize=14, fontweight="bold")

top_countries = cf[country_col].value_counts().head(10)
axes[0].barh(top_countries.index, top_countries.values,
             color=PALETTE["primary"], alpha=0.85)
axes[0].set_title("Customers by Country")
axes[0].set_xlabel("Count")
axes[0].invert_yaxis()

gender_counts = cf[gender_col].value_counts()
axes[1].pie(gender_counts.values, labels=gender_counts.index, autopct="%1.1f%%",
            colors=[PALETTE["primary"], PALETTE["accent"]])
axes[1].set_title("Gender Distribution")

cf["Age"].dropna().hist(ax=axes[2], bins=20,
                        color=PALETTE["secondary"], edgecolor="white", alpha=0.85)
axes[2].set_title("Age Distribution")
axes[2].set_xlabel("Age (years)")
axes[2].axvline(cf["Age"].median(), color=PALETTE["danger"],
                linestyle="--", linewidth=1.5, label=f"Median={cf['Age'].median():.0f}")
axes[2].legend()

plt.tight_layout()
plt.show()

# Active vs inactive customers
active_customers = set(merged["CustomerKey"].unique())
total_customers  = set(cf["CustomerKey"].unique())
inactive         = total_customers - active_customers
print(f"Total Customers     : {len(total_customers):,}")
print(f"Active (have orders): {len(active_customers):,}")
print(f"Inactive (no orders): {len(inactive):,} "
      f"({len(inactive)/len(total_customers)*100:.1f}%) — reactivation opportunity")

# =============================================================================
# STEP 11 — PRODUCT ANALYSIS
# =============================================================================
hdr("STEP 11 — PRODUCT ANALYSIS", 1)

pf2 = FRAMES["products"].copy()
pf2["Margin_pct"] = ((pf2["Unit Price USD"] - pf2["Unit Cost USD"]) /
                      pf2["Unit Price USD"] * 100).round(2)

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle("Product Analysis", fontsize=14, fontweight="bold")

pf2["Unit Price USD"].hist(ax=axes[0], bins=30,
                           color=PALETTE["primary"], edgecolor="white", alpha=0.85)
axes[0].set_title("Unit Price Distribution")
axes[0].set_xlabel("Price ($)")

cat_counts = pf2["Category"].value_counts()
cat_counts.plot(kind="bar", ax=axes[1],
                color=PALETTE["secondary"], edgecolor="white", alpha=0.85)
axes[1].set_title("Products by Category")
axes[1].set_ylabel("SKU Count")
plt.setp(axes[1].get_xticklabels(), rotation=30, ha="right")

margin_by_cat = pf2.groupby("Category")["Margin_pct"].mean().sort_values(ascending=False)
margin_by_cat.plot(kind="bar", ax=axes[2],
                   color=PALETTE["accent"], edgecolor="white", alpha=0.85)
axes[2].set_title("Avg Margin% by Category")
axes[2].set_ylabel("Margin %")
plt.setp(axes[2].get_xticklabels(), rotation=30, ha="right")

plt.tight_layout()
plt.show()

# =============================================================================
# STEP 12 — STORE ANALYSIS
# =============================================================================
hdr("STEP 12 — STORE ANALYSIS", 1)

stf = FRAMES["stores"].copy()

# Revenue per store
store_rev = merged.groupby("StoreKey").agg(
    Revenue=("Revenue","sum"),
    Orders =("Order Number","nunique"),
    Units  =("Quantity","sum"),
).reset_index()
store_full = stf.merge(store_rev, on="StoreKey", how="left")
store_full["Revenue_per_sqm"] = (store_full["Revenue"] /
                                  store_full["Square Meters"]).round(2)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))
fig.suptitle("Store Performance", fontsize=14, fontweight="bold")

by_country = store_full.groupby("Country")["Revenue"].sum().sort_values(ascending=False)
axes[0].bar(by_country.index, by_country.values,
            color=PALETTE["primary"], edgecolor="white", alpha=0.85)
axes[0].set_title("Revenue by Country")
axes[0].set_ylabel("Revenue ($)")
axes[0].yaxis.set_major_formatter(mticker.FuncFormatter(lambda x,_: f"${x/1e6:.1f}M"))
plt.setp(axes[0].get_xticklabels(), rotation=30, ha="right")

sqm = store_full.dropna(subset=["Square Meters","Revenue_per_sqm"])
sqm = sqm[sqm["StoreKey"] != 0]  # exclude online
axes[1].scatter(sqm["Square Meters"], sqm["Revenue_per_sqm"],
                c=PALETTE["secondary"], alpha=0.7, s=60)
axes[1].set_title("Store Size vs Revenue/sqm (Physical Stores)")
axes[1].set_xlabel("Store Size (sqm)")
axes[1].set_ylabel("Revenue per sqm ($)")

plt.tight_layout()
plt.show()

hdr("Top 10 Stores by Revenue", 3)
top_stores = store_full.sort_values("Revenue", ascending=False).head(10)
top_stores["Revenue"] = top_stores["Revenue"].map("${:,.0f}".format)
display(top_stores[["StoreKey","Country","State","Square Meters",
                     "Revenue","Orders","Revenue_per_sqm"]])

# =============================================================================
# STEP 13 — EXCHANGE RATE ANALYSIS
# =============================================================================
hdr("STEP 13 — EXCHANGE RATE ANALYSIS", 1)

ef = FRAMES["exchange_rates"].copy()
ef["Date"]     = pd.to_datetime(ef["Date"],  errors="coerce")
ef["Exchange"] = pd.to_numeric(ef["Exchange"], errors="coerce")

display(ef["Currency"].value_counts().reset_index()
        .rename(columns={"Currency":"Currency","count":"Records"}))

top_currencies = ef["Currency"].value_counts().head(6).index.tolist()
fig, ax = plt.subplots(figsize=(14, 5))
for cur in top_currencies:
    sub = ef[ef["Currency"] == cur].sort_values("Date")
    ax.plot(sub["Date"], sub["Exchange"], label=cur, linewidth=1.5)
ax.set_title("Exchange Rate Trends vs USD", fontsize=13, fontweight="bold")
ax.set_xlabel("Date"); ax.set_ylabel("Rate vs USD")
ax.legend()
plt.tight_layout()
plt.show()

# =============================================================================
# STEP 14 — FEATURE ENGINEERING IDEAS
# =============================================================================
hdr("STEP 14 — FEATURE ENGINEERING IDEAS", 1)

FEATURES = [
    ("revenue_usd",          "quantity * unit_price_usd / exchange_rate",        "Normalize multi-currency revenue for comparison"),
    ("profit_usd",           "revenue_usd - (quantity * unit_cost_usd)",          "Gross profit per line item"),
    ("profit_margin_pct",    "profit_usd / revenue_usd * 100",                    "Margin per transaction"),
    ("delivery_days",        "delivery_date - order_date",                        "Fulfilment speed KPI"),
    ("customer_age",         "(today - birthday) / 365",                          "Age segmentation in Power BI"),
    ("customer_lifetime_value","SUM(revenue_usd) per customer",                   "Total spend; RFM monetary component"),
    ("order_size",           "SUM(quantity) per order_number",                    "Basket size metric"),
    ("store_revenue_sqm",    "annual_revenue / square_meters",                    "Operational efficiency benchmark"),
    ("days_since_last_order","today - MAX(order_date) per customer",              "RFM recency component"),
    ("rolling_30d_revenue",  "30-day rolling SUM of revenue per store",           "Short-term trend smoothing"),
    ("yoy_growth_pct",       "(this_yr - last_yr) / last_yr",                     "Core executive growth KPI"),
    ("abc_segment",          "A=top 70% revenue SKUs, B=next 20%, C=bottom 10%","Inventory prioritisation"),
]

display(pd.DataFrame(FEATURES, columns=["Feature","Formula/Logic","Why Useful"])
        .style.set_caption("Recommended Engineered Features"))

# =============================================================================
# STEP 15 — DBT MODELLING ARCHITECTURE
# =============================================================================
hdr("STEP 15 — dbt MODELLING ARCHITECTURE", 1)
display(Markdown("""
## 📦 Staging Layer (`models/staging/`)
| Model | Source | Key Transformations |
|-------|--------|---------------------|
| `stg_sales` | Sales.csv | Cast dates, rename cols, trim strings |
| `stg_customers` | Customers.csv | Parse birthday, derive age, standardize country |
| `stg_products` | Products.csv | Strip $, cast price/cost to float, normalize categories |
| `stg_stores` | Stores.csv | Cast square_meters, parse open_date |
| `stg_exchange_rates` | Exchange_Rates.csv | Cast rate, parse date, dedup on (date, currency) |

## 🔗 Intermediate Layer (`models/intermediate/`)
| Model | Depends On | Purpose |
|-------|-----------|---------|
| `int_sales_enriched` | stg_sales + stg_exchange_rates | Join FX rates, compute revenue_usd |
| `int_sales_with_profit` | int_sales_enriched + stg_products | Add cost, compute profit_usd, margin% |
| `int_customer_orders` | int_sales_with_profit + stg_customers | CLV, order count, recency per customer |
| `int_store_metrics` | int_sales_with_profit + stg_stores | Revenue, orders, revenue/sqm per store |

## 🏪 Mart Layer (`models/marts/`)
| Model | Purpose |
|-------|---------|
| `mart_sales_fact` | Final enriched fact table (USD normalized) |
| `mart_dim_customer` | Customer master + CLV + RFM segment |
| `mart_dim_product` | Product master + margin + ABC class |
| `mart_dim_store` | Store master + performance metrics |
| `mart_dim_date` | Full date spine for time intelligence |
| `mart_executive_summary` | Pre-aggregated monthly KPIs |
"""))

# =============================================================================
# STEP 16–18 — POWER BI, KPIs, DAX
# =============================================================================
hdr("STEP 16 — POWER BI DASHBOARD PLAN", 1)
display(Markdown("""
| Page | Purpose | Key Visuals |
|------|---------|-------------|
| Executive Dashboard | C-suite overview | KPI cards, revenue trend, category bar, country map |
| Sales Analytics | Transaction patterns | Month×Category matrix, waterfall MoM, scatter AOV vs volume |
| Customer Analytics | Segmentation, loyalty | RFM scatter, age/country charts, CLV table |
| Product Performance | SKU and margin analysis | Top 20 bar, margin scatter, category treemap |
| Store Performance | Geographic benchmarking | Revenue/sqm bar, country map, store heatmap |
| Currency & FX | Multi-currency exposure | FX trend lines, local vs USD revenue bar |

**Global Slicers on all pages:** Year, Country, Category
"""))

hdr("STEP 17 — KPI CATALOGUE", 1)
KPIS = [
    ("Total Revenue USD",       "SUM(revenue_usd)",                     "Top-line metric"),
    ("Gross Profit",            "SUM(profit_usd)",                      "Profitability after COGS"),
    ("Gross Margin %",          "profit / revenue * 100",               "Efficiency benchmark"),
    ("Avg Order Value",         "Total Revenue / Distinct Orders",      "Basket health"),
    ("Units Sold",              "SUM(quantity)",                        "Volume indicator"),
    ("Repeat Purchase Rate",    "Customers >1 order / total",           "Loyalty signal"),
    ("YoY Revenue Growth %",    "(CY-PY)/PY",                          "Growth narrative"),
    ("Store Revenue / sqm",     "store revenue / square meters",        "Operational efficiency"),
    ("Avg Customer Age",        "AVG(age)",                             "Demographic insight"),
    ("Inactive Customer Rate",  "no-order customers / total",           "Reactivation opportunity"),
]
display(pd.DataFrame(KPIS, columns=["KPI","Formula","Business Meaning"])
        .style.set_caption("KPI Catalogue"))

hdr("STEP 18 — DAX MEASURES (Power BI)", 1)
display(Markdown("""
```dax
Total Revenue   = SUM(mart_sales_fact[revenue_usd])
PY Revenue      = CALCULATE([Total Revenue], SAMEPERIODLASTYEAR(mart_dim_date[Date]))
YoY Growth %    = DIVIDE([Total Revenue] - [PY Revenue], [PY Revenue], 0)
Running Total   = CALCULATE([Total Revenue], FILTER(ALL(mart_dim_date), mart_dim_date[Date] <= MAX(mart_dim_date[Date])))
Avg Order Value = DIVIDE([Total Revenue], DISTINCTCOUNT(mart_sales_fact[order_number]))
Gross Profit    = SUM(mart_sales_fact[profit_usd])
Gross Margin %  = DIVIDE([Gross Profit], [Total Revenue], 0)
Revenue/Customer= DIVIDE([Total Revenue], DISTINCTCOUNT(mart_sales_fact[customer_key]))
MoM Growth %    = VAR cm = [Total Revenue]
                  VAR pm = CALCULATE([Total Revenue], DATEADD(mart_dim_date[Date], -1, MONTH))
                  RETURN DIVIDE(cm - pm, pm, 0)
```
"""))

# =============================================================================
# STEP 19 — FINAL SUMMARY REPORT
# =============================================================================
hdr("STEP 19 — FINAL SUMMARY REPORT", 1)
display(Markdown(f"""
| Dimension | Finding |
|-----------|---------|
| **Total Revenue** | ${total_rev:,.0f} across 2016–2021 |
| **Peak Year** | 2019 (~$18M) — COVID caused ~47% decline in 2020 |
| **Top Category** | Computers (34.6% of revenue) — concentration risk |
| **Highest Margin Category** | Music/Movies/Audio Books (57.15%) — undermonetized |
| **Avg Gross Margin** | 54.9% — healthy but suspiciously uniform (formulaic pricing) |
| **Data Quality** | One HIGH issue: Delivery Date 79.1% null (online channel by design) |
| **FK Integrity** | Perfect — 0 orphan records across all 3 FK relationships |
| **Inactive Customers** | ~3,379 customers with no orders — reactivation target |
| **Pareto** | Top 20% SKUs drive 70.3% revenue — build ABC classification |
| **Price Fix Needed** | Unit Cost/Price USD had $ signs — fixed in Step 1b |
| **Composite PK** | Sales grain = (Order Number + Line Item) |
| **Currency** | 5 currencies (USD, CAD, AUD, EUR, GBP) — all covered in exchange_rates |

**Next Steps:**
1. `dbt init retail_bi` → build 5 staging models
2. Build `int_sales_enriched` with FX join for true USD revenue
3. Build `mart_dim_date` date spine for Power BI time intelligence
4. Implement RFM in `int_customer_rfm`
5. Connect Power BI Desktop to dbt mart layer
"""))

# =============================================================================
# STEP 20 — BONUS: CORRELATION + RFM FRAMEWORK + OUTLIERS
# =============================================================================
hdr("STEP 20 — BONUS ADVANCED ANALYSIS", 1)

# Correlation heatmap
hdr("Correlation Matrix — Numeric Columns", 3)
num_cols = merged.select_dtypes(include=[np.number]).columns.tolist()
num_cols = [c for c in num_cols if c not in ["Order Number","CustomerKey",
            "StoreKey","ProductKey","SubcategoryKey","CategoryKey","Line Item"]]
if len(num_cols) > 2:
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.heatmap(merged[num_cols].corr(), annot=True, fmt=".2f",
                cmap="coolwarm", center=0, ax=ax, linewidths=0.5)
    ax.set_title("Correlation Matrix", fontsize=13, fontweight="bold")
    plt.tight_layout()
    plt.show()

# Revenue outlier detection
hdr("Revenue Outlier Detection (IQR Method)", 3)
Q1  = merged["Revenue"].quantile(0.25)
Q3  = merged["Revenue"].quantile(0.75)
IQR = Q3 - Q1
outliers = merged[(merged["Revenue"] < Q1 - 1.5*IQR) |
                   (merged["Revenue"] > Q3 + 1.5*IQR)]
print(f"Outlier rows: {len(outliers):,} ({len(outliers)/len(merged)*100:.1f}% of transactions)")
print(f"IQR Bounds  : ${Q1-1.5*IQR:,.2f}  –  ${Q3+1.5*IQR:,.2f}")

fig, ax = plt.subplots(figsize=(12, 3))
ax.boxplot(merged["Revenue"].dropna(), vert=False, patch_artist=True,
           boxprops=dict(facecolor=PALETTE["primary"], alpha=0.6))
ax.set_title("Revenue per Line Item — Boxplot (IQR outlier bounds)", fontweight="bold")
ax.set_xlabel("Revenue ($)")
plt.tight_layout()
plt.show()

# RFM Framework
hdr("RFM Segmentation Framework", 3)
display(Markdown("""
| Segment | R | F | M | Strategy |
|---------|---|---|---|----------|
| Champions | 5 | 5 | 5 | Reward, upsell premium products |
| Loyal | 4–5 | 4–5 | 3–5 | Loyalty programme, early access |
| At Risk | 2–3 | 3–4 | 3–4 | Win-back campaign, discount offer |
| Hibernating | 1–2 | 1–2 | 1–2 | Reactivation email series |
| Lost | 1 | 1 | 1 | Last-chance or sunset |

**dbt model `int_customer_rfm`:**
```sql
SELECT
    customer_key,
    DATEDIFF('day', MAX(order_date), CURRENT_DATE)  AS recency_days,
    COUNT(DISTINCT order_number)                    AS frequency,
    SUM(revenue_usd)                                AS monetary_usd,
    NTILE(5) OVER (ORDER BY recency_days DESC)      AS r_score,
    NTILE(5) OVER (ORDER BY frequency)              AS f_score,
    NTILE(5) OVER (ORDER BY monetary_usd)           AS m_score
FROM mart_sales_fact
GROUP BY 1
```
"""))

hdr("✅ FULL ANALYSIS COMPLETE — Ready for dbt + Power BI", 1)
print("All 20 steps executed. Review findings above before proceeding to dbt modelling.")
````
